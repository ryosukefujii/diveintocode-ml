# 曽呂利新左衛門問題                                                            
「Pythonを使ってみよう」の問題で扱った紙を折ると月に届く話に似たものとして、曽呂利新左衛門の逸話があります。これを題材にプログラミングを進めます。
>秀吉から褒美を下される際、何を希望するか尋ねられた新左衛門は、今日は米1粒、翌日には倍の2粒、その翌日には更に倍の4粒と、日ごとに倍の量の米を100日間もらう事を希望した。米粒なら大した事はないと思った秀吉は簡単に承諾したが、日ごとに倍ずつ増やして行くと100日後には膨大な量になる事に途中で気づき、他の褒美に変えてもらった。

## 【問題1】100日目の米粒の数
100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）


### 【問題1】解答


```python
days = 100
rice_sin = [1] # 当日のお米の数
rice_cum = [1] # 累積のお米の数

rice = 1 # 初日(0日目)に貰うお米 
total = 1 # 初日(0日目)の累積　

for i in range(days):
  rice = rice*2
  total += rice
  rice_sin.append(rice)
  rice_cum.append(total)

print(rice_sin)
print(rice_cum)

print("{}日目".format(days))
print("もらう米の数： {}粒".format(rice_sin[-1]))
print("累計： {}粒".format(rice_cum[-1]))

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set()
plt.title("Sorori Shinzaemon Question")
plt.xlabel("days")
plt.xticks()
plt.ylabel("number of rice")
plt.plot(rice_sin) 
plt.plot(rice_cum)

plt.show()

```

```python
TOTAL_DAYS = 121 # コースを受講する期間 = 4ヶ月
N_PERSONS = 13 # クラスの受講者数
NEED_CAL = 2200 # 成人男性が1日に必要なカロリー：2200kcal

def check_day(days):
  rice_sin = [1] # 当日のお米の数
  rice_cum = [1] # 累積のお米の数
  rice = 1 # 初日(0日目)の貰うお米 
  total = 1 # 初日(0日目)の累積　
  for i in range(days):
    rice = rice*2
    total += rice
    rice_sin.append(rice)
    rice_cum.append(total)
  return rice_cum[-1]


def cal_sorori_shizaemon(n_rice, n_persons):
    RICE_VOLUME = 0.044 # 炊いた米1粒：0.44g ← 炊く前×吸水による増加率 = 0.02g * 2.2
    RICE_CAL = 0.0336 # 米1粒あたりのカロリー：0.0336kcal(ご飯100gあたり:168kcal → ご飯1gあたり:1.68kcal) 
    total_cal = n_rice * RICE_CAL / n_persons # 1人当たりの総カロリー
    liveable_days = total_cal / NEED_CAL # 生活できる日数
    print("生活できる日数： {:.1f}日".format(liveable_days))
    print("1人当たりの総カロリー： {:.0f}kcal".format(total_cal))
   

cal_sorori_shizaemon(check_day(TOTAL_DAYS), N_PERSONS)
```
