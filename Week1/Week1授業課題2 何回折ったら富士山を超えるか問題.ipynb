# 何回折ったら富士山を超えるか問題
「Pythonを使ってみよう」の問題では「1枚の紙を43回折り曲げた時の厚さを計算するプログラム」を作成し、43回折ることで月に届くことが確認できました。

ここで「何回折ったらある厚さになるのか」という疑問が生まれます。こういったことを計算するプログラムを作成していきましょう。

### 紙を折った時の厚さの数式
授業前課題と同様ですが確認しておきます。
n回折った紙の厚さ $t_{n}$は以下のように表せます。


$
t_{n} = t_{0}×2^{n}
$

折る前の紙の厚さ$t_{0}$は**0.00008m**とします。一般的なコピー用紙の厚さです。

##【問題1】富士山を超える最小の折る回数 
厚さが「富士山(3776m)」を超えるのは何回紙を折った時か計算するプログラムを作成してください。

### 【問題1】解答


```python
THICKNESS = 0.00008 # 紙の厚さ
height = THICKNESS
cnt = 0 # 折った回数

mt_fuji = 3776 # 富士山までの距離：384,400kmをメートルに変換

while mt_fuji > height: 
    height = height * 2
    cnt += 1

print('折った回数：{}回'.format(cnt))
```

##【問題2】任意の厚さに対応した関数を作成
距離を入力すると、何回折った時に超えるかを出力する関数を作成してください。引数として、折る前の厚さ$t_{0}$も設定できるようにしましょう。 

この関数を利用して、「最も近い太陽以外の恒星」に到達するには紙を何回折る必要があるか計算してください。「最も近い太陽以外の恒星」の距離は調査してください。

### 【問題2】解答


```python
def fold_paper(thickness=0.00008, distance=None):
    cnt = 0 # 折った回数
    while  distance > thickness : 
        thickness = thickness * 2
        cnt += 1    
    return cnt

# 最も近い太陽以外の恒星 = プロキシマ・ケンタウリ
pc_dis = 4.243 * 9.46 * 10**15 # プロキシマ・ケンタウリから地球までの距離： 4.243光年をメートルに変換
print('折った回数：{}回'.format(fold_paper(distance=pc_dis)))
```

##【問題3】必要な紙の長さ
 実際のところ身近にある紙は43回はおろか10回も折れません。しかし、大きな紙ほどたくさん折ることができ、トイレットペーパーのような長い紙を一方向に折ることを考えた時の折れる回数を求める公式が存在します。
 
厚さ$t_{0}$の紙をn回折るために必要な紙の長さLは以下の公式で求まります。

$ L = \frac{\pi t_{0}}{6}(2^n + 4)(2^n - 1) $

この公式を利用して、実際に紙を折り任意の厚さに到達するために必要な紙の長さを求める関数を作成してください。
そしてこれを使って「月」「富士山」「最も近い太陽以外の恒星」に届くために必要な紙の長さを求めてください。

**《公式の出典》**

[Folding Paper in Half | Teachers of India \](http://teachersofindia.org/en/ebook/folding-paper-half)


**《解説記事》**

[ Britney Gallivan - Wikipedia](https://en.wikipedia.org/wiki/Britney_Gallivan)


**《ヒント》**

この公式の中で折る回数nが登場しますが、このnは先ほど作成した関数で求まります。関数の中で関数を使うことが可能です。

### 【問題3】解答

```python
import numpy as np


def paper_length(thickness=0.00008, n_folded=None):
    length = (np.pi * thickness / 6) * (2**n_folded + 4) * (2**n_folded -1)
    return length

THICKNESS = 0.00008 # 紙の厚さ
height = THICKNESS
cnt = 0 # 折った回数

# 月
moon = 384400000 # 月までの距離：384,400kmをメートルに
moon_folded = fold_paper(distance=moon) 
print('折った回数：{:.1f}回'.format(paper_length(n_folded=moon_folded)))

# 富士山
mt_fuji = 3776 # 富士山までの距離：384,400kmをメートルに変換
mt_fuji_folded = fold_paper(distance=mt_fuji) 
print('折った回数：{:.1f}回'.format(paper_length(n_folded=mt_fuji_folded)))


# 最も近い太陽以外の恒星 = プロキシマ・ケンタウリ
pc_dis = 4.243 * 9.46 * 10**15 # プロキシマ・ケンタウリから地球までの距離： 4.243光年をメートルに変換
pc_dis_folded = fold_paper(distance=pc_dis) 
print('折った回数：{:.1f}回'.format(paper_length(n_folded=pc_dis_folded)))


```
