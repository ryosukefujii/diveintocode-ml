# ダミーデータの作成
機械学習に関するプログラムの実験を行う上で、実データではないランダムなデータ（ダミーデータ）が欲しいということがあります。正規分布に従う乱数によるダミーデータの作成を通じてNumPyの使い方になれていきましょう。

## 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

$$
% <![CDATA[
\left[
\begin{array}{ccc}
  1.0 & 0.8\\
  0.8 & 1.0
\end{array}
\right] %]]>
$$

多次元の正規分布による乱数は**np.random.multivariate_normal()**を使うことで作成できます。

[numpy.random.multivariate_normal — NumPy v1.16 Manual
](https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.multivariate_normal.html)

作成したndarrayのshapeは(500, 2)になります。

### 《シード値について》

実行ごとに値が変わらないように、np.random.seed()により乱数のシード値を固定しておきましょう。今回シード値は0を使うことにします。

```
np.random.seed(0)
```

### 【問題1】解答


````python
import numpy as np

np.random.seed(0)

samples = 500 # サンプル数
mu = [-3, 0] # 平均
sigma = [[1.0, 0.8], [0.8, 1.0]] # 共分散列

# 2次元正規乱数をサンプルの数だけ生成
values = np.random.multivariate_normal(mu, sigma, samples)

print(values)
````

````python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns

np.random.seed(0)

samples = 500 # サンプル数
mu = [-3, 0] # 平均
sigma = [[1.0, 0.8], [0.8, 1.0]] # 共分散列

# 2次元正規乱数をサンプルの数だけ生成
values = np.random.multivariate_normal(mu, sigma, samples)

sns.set()
plt.title("Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(values[:, 0], values[:, 1])
plt.show()
````

### 【問題3】解答

````python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns

np.random.seed(0)

samples = 500 # サンプル数
mu = [-3, 0] # 平均
sigma = [[1.0, 0.8], [0.8, 1.0]] # 共分散列

# 2次元正規乱数をサンプルの数だけ生成
values = np.random.multivariate_normal(mu, sigma, samples)

# 上限、下限設定
min_value = values.min()
max_value = values.max()

sns.set()
plt.title("Histogram x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(values[:, 0]) 
plt.xlim([min_value, max_value])
plt.show()

plt.title("Histogram of x2")
plt.xlabel("x2")
plt.ylabel("frequency")
plt.hist(values[:, 1])
plt.xlim([min_value, max_value])
plt.show()
````

## 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。

0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。

機械学習に使用するデータセットはこのような形になっていることが多いです。

**《作成例》**

データが3個+3個の場合は以下のような(6, 3)のndarrayになります。

![代替テキスト](https://t.gyazo.com/teams/diveintocode/a705e6e96f79e636934103088de0a0d0.png)

### 【問題6】解答

````python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns

np.random.seed(0)

samples = 500 # サンプル数
mu = [-3, 0] # データ1平均
sigma = [[1.0, 0.8], [0.8, 1.0]] # 共分散列

# データ1
label_zero = np.array([0]*samples).reshape(500, 1)
data1 = np.concatenate([np.random.multivariate_normal(mu, sigma, samples), label_zero], 1)

# データ2
mu2 = [0, -3] # データ2平均
label_one = np.array([1]*samples).reshape(500, 1)
data2 = np.concatenate([np.random.multivariate_normal(mu2, sigma, samples), label_one], 1)

data3 = np.concatenate([data1, data2], 0)
print(data3.shape)
print(data3)
````
